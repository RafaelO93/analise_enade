```python
import pandas as pd
import numpy as np
import math
```

```python
def switch_QE_I08(chave):
    switcher = {
        "A": "Até 1,5 sálarios-minimos",
        "B": "1,5 a 3 sálarios-minimos",
        "C": "3 a 4,5 sálarios-minimos",
        "D": "4,5 a 6 sálarios-minimos",
        "E": "6 a 10 sálarios-minimos",
        "F": "10 a 30 sálarios-minimos",
        "G": "Acima de 30 sálarios",
        "NI": "Não informado"
    }
    return switcher[chave]
```

```python
def switch_QE_I11(valor):
    nenhum = ["A", "B"]
    parcial = ["D", "F"]
    if valor in nenhum:
        return "Nenhum investimento"
    elif valor in parcial:
        return "Investimento Parcial"
    elif valor=="NI":
        return "Não Informado"
    else:
        return "Financiamento total"
```

```python
dados = pd.read_csv("dados/enade2017_computacao_nordeste.csv")
dados["QE_I08"] = dados["QE_I08"].fillna("NI")
dados["QE_I11"] = dados["QE_I11"].fillna("NI")
dados.shape[0]
```

```python
result = []
lista_valores = list(dados["QE_I08"])
for i in range(len(lista_valores)):
    result.append(switch_QE_I08(lista_valores[i]))
    
result2 = []
lista_valores2 = list(dados["QE_I11"])
for i in range(len(lista_valores2)):
    result2.append(switch_QE_I11(lista_valores2[i])) 
```

```python
dados["QE_I08_Convertido"] = result
dados["QE_I11_Convertido"] = result2
dados.shape[0]
```

```python
dados_conceito = pd.read_excel("dados/resultados_conceito_enade_2017.xlsx")
```

```python
dados_conceito["CO_IES"] = dados_conceito[dados_conceito.columns[3]]
dados_conceito["CO_CURSO"] = dados_conceito[dados_conceito.columns[8]]
dados_conceito["CONCEITO"] = dados_conceito[dados_conceito.columns[20]]
dados_conceito = dados_conceito[["CO_IES", "CO_CURSO", "CONCEITO"]]
```

```python
final = pd.merge(dados, dados_conceito, left_on=['CO_IES', 'CO_CURSO'], right_on=['CO_IES', 'CO_CURSO'], how='inner')
```

```python
final.to_csv("dados/enade2017_computacao_nordeste_convertido.csv", index=False, header=True)
final = pd.read_csv("dados/enade2017_computacao_nordeste_convertido.csv")
```
